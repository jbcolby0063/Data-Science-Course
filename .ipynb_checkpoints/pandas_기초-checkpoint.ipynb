```python
# method 1: list
two_dimensional_list = [['dongwook', 50, 86], ['sineui', 89, 31], ['ikjoong', 68, 91], ['yoonsoo', 88, 75]]
# method 2: np array
two_dimensional_array = np.array(two_dimensional_list)
# method 3: pd Series
list_of_series = [
    pd.Series(['dongwook', 50, 86]), 
    pd.Series(['sineui', 89, 31]), 
    pd.Series(['ikjoong', 68, 91]), 
    pd.Series(['yoonsoo', 88, 75])
]

# 아래 셋은 모두 동일합니다
df1 = pd.DataFrame(two_dimensional_list)
df2 = pd.DataFrame(two_dimensional_array)
df3 = pd.DataFrame(list_of_series)

df1
```

```python
names = ['dongwook', 'sineui', 'ikjoong', 'yoonsoo']
english_scores = [50, 89, 68, 88]
math_scores = [86, 31, 91, 75]

# method 1: dict
dict1 = {
    'name': names, 
    'english_score': english_scores, 
    'math_score': math_scores
}
# method 2: np array dict
dict2 = {
    'name': np.array(names), 
    'english_score': np.array(english_scores), 
    'math_score': np.array(math_scores)
}
# method 3: pd Series dict
dict3 = {
    'name': pd.Series(names), 
    'english_score': pd.Series(english_scores), 
    'math_score': pd.Series(math_scores)
}


# 아래 셋은 모두 동일합니다
df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
df3 = pd.DataFrame(dict3)

df1
```

```python
# row에서 디스플레이가 5보다 큰 애들만 출력
iphone_df.loc[iphone_df['디스플레이'] > 5]
iphone_df[iphone_df['디스플레이'] > 5] # .loc[] 없이도 가능

# 주의! .loc[]은 input으로 DataFrame이 허용 x
# 그래서 'iphone_df['디스플레이', '메모리']'나 'iphone_df.loc[:, '디스플레이':'메모리']'같은 DataFrame을 input으로 사용하고 싶을 때, 그냥 df[] 사용 
# - iphone_df.loc[iphone_df[['디스플레이', '메모리']] > 5] --> Error (x)
# - iphone_df[iphone_df[['디스플레이', '메모리']] > 5] --> (o)
# - iphone_df[iphone_df.loc[:, '디스플레이':'메모리'] > 5] --> (o)
```

```python
# row에서 Face ID가 되는 애들만 출력
iphone_df.loc[iphone_df['Face ID'] == 'Yes']
```

```python
# row에서 디스플레이가 5보다 크고 Face ID가 되는 애들만 출력
condition = (iphone_df['디스플레이'] > 5) & (iphone_df['Face ID'] == 'Yes')
iphone_df.loc[condition]
```

```python
# row에서 디스플레이가 5보다 크거나 Face ID가 되는 애들만 출력
condition = (iphone_df['디스플레이'] > 5) | (iphone_df['Face ID'] == 'Yes')
iphone_df.loc[condition]
```

```python
# 위 조건을 만족시키는 row에서 출시일만 출력
iphone_df.loc[condition, '출시일']
```

```python
iphone_df = pd.read_csv('data/iphone.csv', index_col=0)
print("변경 전: ")
print(iphone_df)

# 조건식으로 변경 (1-1)
# df[]나 df.loc[]에 1차원 배열이나 pandas Series가 input으로 들어가면 조건을 만족시키는 row의 모든 원소 변경
iphone_df.loc[iphone_df['디스플레이'] > 5] = 'p'
print("\n변경 후: ")
print(iphone_df)
```

```python
iphone_df = pd.read_csv('data/iphone.csv', index_col=0)
print("변경 전: ")
print(iphone_df)

# 조건식으로 변경 (1-2)
# 조건을 만족시키는 row에서 지정된 column의 값만 변경
iphone_df.loc[iphone_df['디스플레이'] > 5, '출시 버전'] = 'r'
print("\n변경 후: ")
print(iphone_df)
```

```python
iphone_df = pd.read_csv('data/iphone.csv', index_col=0)
iphone_df.loc['iPhone X', '디스플레이'] = 'p'
iphone_df.loc['iPhone XS', '디스플레이'] = 'p'
iphone_df.loc['iPhone XS', '메모리'] = 'p'
iphone_df.loc['iPhone XS Max', '메모리'] = 'p'
print("변경 전: ")
print(iphone_df)

# 조건식으로 변경 (2)
# df[]에 2차원 배열이나 pandas DataFrame이 input으로 들어가면 조건을 만족시키는 원소만 변경
iphone_df[iphone_df.loc[:, '디스플레이':'메모리'] == 'p'] = 'q'
print("\n변경 후: ")
print(iphone_df)
```
